#### Difference between Pandas Series and DataFrame

> A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects.
>
>A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.).
>
> A Series only has one data column, whereas a single-column DataFrame contains both a data column and an index. The labels for the data are contained in the index, which is a second column.

```python
# You can create a new column by using the filter condition
# df.loc[condition, new column name] = 'value if condition is true'

# If Art is greater than or equal to 90, create new column Art_Grade and assign Grade 'A'
# Note all other values will have a value of NaN = Not a Number
# This tells that we have not assigned any values to this column

df.loc[df['Art'].ge(90), 'Art_Grade'] = 'A'

df
```

```python
# We can check for all rows that have been assigned values using .notnull() condition

df[df['Art_Grade'].notnull()].head(10)
```

```python
# To see all the values that do not have any values assigned, you can use .isnull()

df[df['Art_Grade'].isnull()].head(10)
```

```python
# We can reuse the same column and add more conditions and replace values for the NaN

# If Art is between 80 and 89 inclusive, create new column Art_Grade and assign Grade 'B'
# If column already exists, use the same column

df.loc[df['Art'].between(80,89),'Art_Grade'] = 'B'

# If Art is between 70 and 79 inclusive, create new column Art_Grade and assign Grade 'C'
# If column already exists, use the same column

df.loc[df['Art'].between(70,79),'Art_Grade'] = 'C'

# If Art is between 60 and 69 inclusive, create new column Art_Grade and assign Grade 'D'
# If column already exists, use the same column

df.loc[df['Art'].between(60,69),'Art_Grade'] = 'D'

# If Art is less than or equal to 59, create new column Art_Grade and assign Grade 'F'
# If column already exists, use the same column

df.loc[df['Art'].le(59),'Art_Grade'] = 'F'

df
```

```python
# The above method although a good one, is not recommended. 
# Instead create a function to calculate the value
# Then call the function using .apply(function)
# Make sure function returns the value you need to assign back to new column

def grade_check(x):
    if x <= 59:
        return 'F'
    elif x in range(60,70):
        return 'D'
    elif x in range(70,80):
        return 'C'
    elif x in range(80,90):
        return 'B'
    elif x >= 90:
        return 'A'
    else:
        return 'N/A'
```

```python
# You can create a new column using left hand side assignment. 
# Then call .apply() method on the column of choice 
# .apply will call the function. This will return a value which then gets assigned to each row

df['English_Grade'] = df['English'].apply(grade_check)

df.head(10)
```

```python
# For the other two columns, we can use a loop and process them one at a time
# Same concept but using a for loop

temp = ['Math','Science']

for t in temp:
    temp_col = t + '_Grade'
    df[temp_col] = df[t].apply(grade_check)

# Let's see what the columns and values look like now
df.head(10)
```

```python
# Let's look at the columns and their Data Types for the newly created columns

df.info()
```

```python
# You can get the list of all the columns using df.columns

df.columns
```

#### Summary of the MERGE Pandas statements (equivalent SQL JOIN statements)



```python
# Let's first create 2 Pandas DataFrames - One Left and the other Right
# First DataFrame will be df_left and defined as follows:

df_left = pd.DataFrame({'key'   :   ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'], 
                        'value' : [  4,   2,   1,   7,   8,   6,   3,   5]})
df_left
```

```python
# Second DataFrame is df_right and defined as follows:

df_right = pd.DataFrame({'key'   : ['B', 'D', 'E', 'F', 'K', 'L' , 'M', 'N'],
                         'value' : [  2,   4,   3,   1,   6,    7,   5,   8]})
df_right
```

#### Let's do a INNER JOIN on the two DataFrames  (option: how='inner')
> This will merge the data to the left with the right (intersection only)
> - blue indicates rows that are present in the merge result
> - red indicates rows that are excluded from the result (i.e., removed)

<div style="max-width:300px;margin-left: 100">
  
</div>

```python
# INNER JOIN will only pick the common rows in df_left and df_right
# we will get rows B, D, E, and F. All others will be excluded

df_inner = df_left.merge(df_right, on='key', how='inner')

df_inner
```

#### Let's do a LEFT OUTER JOIN on the two DataFrames (option: how='left')
> This will keep all data on the left and merge only matching data on the right 
> - blue indicates rows that are present in the merge result
> - red indicates rows that are excluded from the result (i.e., removed)
> - green indicates rows that are included but missing values are marked NaN
<div style="max-width:300px;margin-left: 100">
  
</div>

```python
# LEFT OUTER JOIN will pick all rows from df_left and the common rows in df_right
# we will get rows A, B, C, D, E, F, G, and H. All others will be excluded
# Note that any row that does not exist in right will be assigned NaN for value_y column
# Rows K, L, M, and N are ignored

df_left_outer = df_left.merge(df_right, on='key', how='left')

df_left_outer
```

#### Let's do a RIGHT OUTER JOIN on the two DataFrames (option: how='right')
> This will keep all data on the right and merge only matching data on the left 
> - blue indicates rows that are present in the merge result
> - red indicates rows that are excluded from the result (i.e., removed)
> - green indicates rows that are included but missing values are marked NaN
<div style="max-width:300px;margin-left: 100">
  
</div>

```python
# RIGHT INNER JOIN will pick all rows from df_right and the common rows in df_left
# we will get rows B, D, E, F, K, L, M, and N. All others will be excluded
# Note that any row that does not exist in left will be assigned NaN for value_x column
# Rows A, C, G, and H are ignored
# For this, use option : how='right'

df_right_outer = df_left.merge(df_right, on='key', how='right')

df_right_outer
```

#### Let's do a FULL OUTER JOIN on the two DataFrames (option: how='outer')
> This will keep all data on the right and merge only matching data on the left 
> - blue indicates rows that are present in the merge result
> - green indicates rows that are included but missing values are marked NaN
<div style="max-width:300px;margin-left: 100">
  
</div>

```python
# FULL OUTER JOIN will pick all rows from df_left and rows from df_right
# we will get rows A, B, C, D, E, F, G, H, K, L, M and N. All others will be excluded
# Note that all rows will be added. 
# The rows that do not match will have NaN for value_x and value_y column
# For this, use option : how='outer'

df_right_outer = df_left.merge(df_right, on='key', how='outer')

df_right_outer
```

## Pandas Groupby 

#### Quick overview
Pandas has multiple options for Groupby. We will just use a simple Groupby 
to capture the data and plot the values on a bar chart

To use Pandas Groupby, the format is:

`df.groupby([groupby_col_list])[math_col_list].func()`

groupby_col_list is the list of columns you want to group by. 
If more than one, Pandas will create a multi-index column

math_col_list is the column that needs to be processed for mathematical operation
math_col_list can be numeric or non-numeric. If numeric, we can use statistical operations
If non-numeric, only count, max, min and other string operations can be performed
If math_col_list is left out, Pandas will do mathematical operations on all other columns

func() is the function that will be done on the math_col_list columns.
Some of the examples of functions are: mean(), min(), max(), avg(), sum(), count()
You can also give custom function using .apply() to call your own function

```python
df_Grades[['Art_Grade']].sort_values('Art_Grade')
```

```python
# Let's count the number of Students that has Letter Grades of A, B, C, D, and F
df_Art = df_Grades.groupby(['Art_Grade'])['Grade_Level'].count()
df_Art
```

```python
# Let's count the number of Students in each Grade Level
df_Grade_Level = df_Grades.groupby(['Grade_Level']).count()
df_Grade_Level
```

```python
# Let's find out how the Students did in Art Class. Let's group by
# Grade Level and then count the letter grades of Art Class. 
# The Letter Grades are A, B, C, D, F

df_Art_by_Grades = df_Grades.groupby(['Grade_Level','Art_Grade'])['Art_Grade'].count()
df_Art_by_Grades
```
